```
train = DirectoryIterator('/content/drive/MyDrive/ИИ/', ImageDataGenerator(), target_size=(224, 224), color_mode='rgb')
test = DirectoryIterator('/content/drive/MyDrive/ИИ/', ImageDataGenerator(), target_size=(224, 224), color_mode='rgb')
```

```
plt.figure(figsize=(10, 5))
plt.suptitle('Train')
cls_ = { v: k for k, v in train.class_indices.items()}
for idx, img in enumerate(train[0][0][:10]):
    ax = plt.subplot(2, 5, idx + 1)
    plt.imshow(img.astype('int'))
    ax.set_title(cls_[train[0][1][idx].argmax()])
plt.tight_layout()
```

```
plt.figure(figsize=(10, 5))
plt.suptitle('Test')
cls_ = { v: k for k, v in train.class_indices.items()}
for idx, img in enumerate(test[0][0][:10]):
    ax = plt.subplot(2, 5, idx + 1)
    plt.imshow(img.astype('int'))
    ax.set_title(cls_[train[0][1][idx].argmax()])
plt.tight_layout()
```

```
model = Sequential()
model.add(Conv2D(8, 3, input_shape=(train.image_shape)))
model.add(MaxPool2D(2))
model.add(Conv2D(16, 3))
model.add(MaxPool2D(2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(train.num_classes, activation='softmax'))

model.compile(Adam(), categorical_crossentropy, metrics=['acc'])

out = model.fit(train, validation_data=test, epochs=10, verbose=2)
```

```
plt.figure()
plt.title('Loss')
plt.plot(out.history['loss'], label='train')
plt.plot(out.history['val_loss'], label='test')
plt.legend()

plt.figure()
plt.title('Acc')
plt.plot(out.history['acc'], label='train')
plt.plot(out.history['val_acc'], label='test')
plt.legend()
```

```
import keras.applications
import keras.layers
import keras.preprocessing
from keras.applications.resnet_v2 import preprocess_input, ResNet152V2
from keras.layers import Dense, BatchNormalization, GlobalAveragePooling2D, Dropout
from keras.models import Model
from keras.optimizers import Adam
from keras.losses import categorical_crossentropy
```

```
IMG_SIZE = [224, 224]
```

```
train = DirectoryIterator('/content/drive/MyDrive/ИИ/',
                          ImageDataGenerator(preprocessing_function=preprocess_input), target_size=IMG_SIZE, color_mode='rgb')
test = DirectoryIterator('/content/drive/MyDrive/ИИ/',
                         ImageDataGenerator(preprocessing_function=preprocess_input), target_size=IMG_SIZE, color_mode='rgb')
```

```
base_model= ResNet152V2(weights="imagenet",  # Load weights pre-trained on ImageNet.
    input_shape=(*IMG_SIZE, 3),
    include_top=False,)

for layer in base_model.layers:
    layer.trainable = False
```

```
base_model.input_shape
```

```
base_model.output_shape
```

```
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(1280, activation='relu', bias_initializer='zeros')(x)
# x = Dropout(0.5)(x)
x = BatchNormalization()(x)
predictions = Dense(train.num_classes, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)
model.compile(optimizer=Adam(),
              loss=categorical_crossentropy,
              metrics=["acc"])

out = model.fit(train, epochs=5, verbose=1, validation_data=test)
```
